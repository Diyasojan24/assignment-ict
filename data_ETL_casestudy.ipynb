```python
import pandas as pd
```

```python
url="https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv"
```

**Load the dataset into a DataFrame**

```python
df=pd.read_csv(url)
df
```

**Identify number of records and unique locations**

```python
num_records=len(df)
unique_loc=df["place"].nunique()
print("number of records:",num_records)
print("number of unique locations:",unique_loc)
```

**Print top 5 rows and column** **names**

```python
print(df.head())
print(df.columns.tolist())
```

**Convert time to datetime**


```python
df['time']=pd.to_datetime(df['time'])
```

**Drop records with missing latitude, longitude, mag**

```python
df=df.dropna(subset=['latitude','longitude','mag'])
df.head()
```

**Filter only earthquakes with magnitude >= 4.0**

```python
df=df[df['mag']>=4.0]
df.head()
```

**Add a column day_of_week from time**


```python
df['day_of_week']=df['time'].dt.day_name()
df
```

Create a column severity_level based on magnitude:
‚óã <4.0: "Low"
‚óã 4.0 - 6.0: "Moderate"
‚óã 6.0+: "High ***


```python
def classify_severity(mag):
  if mag<4.0:
   return "low"
  elif mag>=4.0 and mag<=6.0:
   return "moderate"
  else:
    return "High"
df['severity_level']=df['mag'].apply(classify_severity)
df.head()
```

**Count number of earthquakes per place**


```python
earthquakes_per_place=df['place'].value_counts()
earthquakes_per_place
```

Compute average magnitude and max depth per day

```python
df['date']=df['time'].dt.date
summary=df.groupby('date').agg(avg_magnitude=('mag','mean'),max_depth=('depth','max')).reset_index()
print(summary.head())
```

Save:
‚óã Cleaned dataset to cleaned_earthquakes.csv **bold text**

```python
print(df.to_csv("cleaned_earthquakes.csv",index=False))
```

```python
print(summary.to_csv("earthquake_summary.csv "))
```
